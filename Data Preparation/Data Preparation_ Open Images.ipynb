## Load a dataset

Let's get started by importing the FiftyOne library:

```python
import fiftyone as fo
```

```python
import fiftyone.zoo as foz

# Filter
classes = ["Backpack", "Mobile phone", "Handbag", "Suitcase"]

dataset = foz.load_zoo_dataset(
              "open-images-v6",
              label_types=["detections"],
              classes=classes,
              only_matching=True,
              num_workers=24,
              # max_samples=1,
              # seed=51,
              # shuffle=True,
              dataset_name="Open-Images-lost-items",
          )
```

```python
print(dataset)
```

# Export the view (filter)


1.   Create a view that only contains filtered classes
2.   Override dataset.default_classes by manually providing the `classes` argument

```python
from fiftyone import ViewField as F

export_dir="Open-Images"
label_field = "detections"
# dataset_type = fo.types.COCODetectionDataset
dataset_type = fo.types.YOLOv5Dataset

# The splits to export
splits = ["train", "test", "validation"]

view = dataset.filter_labels(label_field, F("label").is_in(classes))

for split in splits:
    split_view = view.match_tags(split)
    split_view.export(
        export_dir=export_dir,
        dataset_type=dataset_type,
        label_field=label_field,
        split=split,
        classes=classes,
    )
```
